```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = df_1.drop(['FWI'],axis = 1)
y = df_1['FWI']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=57)
```

```python
def correlation(dataset,threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                col_name = corr_matrix.columns[i]
                col_corr.add(col_name)
    return col_corr
```

```python
corr_features = correlation(X_train,0.85)
corr_features
```

```python
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)

X_train.shape,X_test.shape
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

lin_reg = make_pipeline(
    StandardScaler(),
    LinearRegression()
)

lin_reg.fit(X_train,y_train)
```

```python
lin_preds = lin_reg.predict(X_test)
lin_preds
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
def evaluation_preds(y_test,preds):
    lin_mae = mean_absolute_error(y_test,preds)
    lin_mse = mean_squared_error(y_test,preds)
    lin_rmse = np.sqrt(lin_mse)
    lin_r2 = r2_score(y_test,preds)
    metrics = {'mae':lin_mae,
               'mse':lin_mse,
               'rmse':lin_rmse,
               'R^2':lin_r2}
    return metrics
```

```python
lin_results = evaluation_preds(y_test,lin_preds)
lin_results
```

```python
from sklearn.linear_model import Lasso

lasso_reg = make_pipeline(
    StandardScaler(),
    Lasso()
)

lasso_reg.fit(X_train,y_train)

lasso_preds = lasso_reg.predict(X_test)

lasso_results = evaluation_preds(y_test,lasso_preds)

print(lasso_results)
```

```python
from sklearn.linear_model import LassoCV

lassocv = make_pipeline(
    StandardScaler(),
    LassoCV(cv = 5)
)

lassocv.fit(X_train,y_train)

lassocv_preds = lassocv.predict(X_test)

lassocv_results = evaluation_preds(y_test,lassocv_preds)

print(lassocv_results)
```

## RIDGE REGRESSION

```python
from sklearn.linear_model import Ridge

ridge_reg = make_pipeline(
    StandardScaler(),
    Ridge()
)

ridge_reg.fit(X_train,y_train)

ridge_preds = ridge_reg.predict(X_test)

ridge_results = evaluation_preds(y_test,ridge_preds)

print(ridge_results)

```

```python
from sklearn.linear_model import RidgeCV

ridgecv = make_pipeline(
    StandardScaler(),
    RidgeCV(cv = 5)
)

ridgecv.fit(X_train,y_train)

ridgecv_preds = ridgecv.predict(X_test)

ridge_cv_results = evaluation_preds(y_test,ridgecv_preds)

print(ridge_cv_results)
```

```python
from sklearn.linear_model import ElasticNet

elastic_reg = make_pipeline(
    StandardScaler(),
    ElasticNet()
)

elastic_reg.fit(X_train,y_train)

elastic_preds = elastic_reg.predict(X_test)

elastic_results = evaluation_preds(y_test,elastic_preds)

print(elastic_results)
```

```python
from sklearn.linear_model import ElasticNetCV

elastic_cv = make_pipeline(
    StandardScaler(),
    ElasticNetCV()
)

elastic_cv.fit(X_train,y_train)

elastic_cv_preds = elastic_cv.predict(X_test)

elastic_cv_results = evaluation_preds(y_test,elastic_cv_preds)

print(elastic_cv_results)
```

```python
di = {'lasso':lasso_results,
      'lasso_cv': lassocv_results,
      'ridge':ridge_results,
      'ridge_cv':ridge_cv_results,
      'elasticnet':elastic_results,
      'elastic_cv':elastic_cv_results}

results = pd.DataFrame(di)

results
```
